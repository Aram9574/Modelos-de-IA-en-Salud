# Imputacion de datos y normalizacion

```python
"""Toda esta parte del codigo se usa para elmininar los nulos y escalar los datos"""

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Separar variables caracteristicas y objetivo
"""X siempre corresponde a la variable objetivo, mientras y corresponde a las demas"""

X = df.drop(columns=["TenYearCHD"])#Variale objetivo, debe ser binaria
y = df["TenYearCHD"]#Demas variables

# Imputar faltantes con la media
"""Decido usar la media porque no depende del valor del dato anterior o
posterior como si lo hace con ffill() y bfill(). La media rellena los nulos
con valores globales. no tendria sentido que un paciente de 35 años sano tenga la
presion arterial de un paciente de 80 con hipertension"""

imputer = SimpleImputer(strategy="mean")
X_imp = imputer.fit_transform(X)

# Estandarizar para que todas las variables estén en escala similar
scaler = StandardScaler() #Media de 0 y desviacion de 1
X_scaled = scaler.fit_transform(X_imp)

# Volver a DataFrame con nombres originales
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

display(X_scaled.head())
```

# 4. Seleccion de caracteristicas

```python
"""El metodo KBest selecciona las variables que mas se relacionan con el objetivo
de forma automatica. Pense en agregar totChol porque desde un punto de vista medico
el colesterol tiene relevancia, pero segun el selector automatico, esta variable por
si sola no tiene suficiente peso sobre el resultado final.
Agregarla resulta en una leve disminucion en la sensibilidad del modelo"""

from sklearn.feature_selection import SelectKBest, f_classif

# Elegir variables mas asociadas a objetivo, seleccion de 8 con k
selector = SelectKBest(score_func=f_classif, k=8)
X_sel = selector.fit_transform(X_scaled, y)

cols_sel = X.columns[selector.get_support()].tolist()
print("Características seleccionadas:", cols_sel)

# Dataset reducido final
X_final = pd.DataFrame(X_sel, columns=cols_sel)

print("\nDataset reducido:")
display(X_final.head())
display(X_final.info())
```

```python
"""Toda esta parte es extra, solo la uso para comparar como se ven los resultados
de los modelos si se le hacen ajustes sobre el umbral y el balanceo de clases"""

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Esta funcion evalua el modelo dado un umbral
def evaluar_modelo(modelo, X_test, y_test, umbral=0.5):
    y_prob = modelo.predict_proba(X_test)[:, 1]
    y_pred = (y_prob >= umbral).astype(int)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    return {"Accuracy": acc, "Precision": prec, "Recall": rec, "F1-score": f1, "CM": cm}

# 1. Modelo original sin balancear
modelo_original = LogisticRegression(max_iter=1000)
modelo_original.fit(X_train, y_train)
res_original = evaluar_modelo(modelo_original, X_test, y_test, umbral=0.5)

# 2. Modelo balanceado
modelo_bal = LogisticRegression(max_iter=1000, class_weight='balanced')
modelo_bal.fit(X_train, y_train)
res_bal = evaluar_modelo(modelo_bal, X_test, y_test, umbral=0.5)

# 3. Modelo balanceado + umbral reducido
res_bal_umbral = evaluar_modelo(modelo_bal, X_test, y_test, umbral=0.3)

# 4. Tabla comparativa
df_comp = pd.DataFrame({
    "Modelo original": res_original,
    "Balanceado": res_bal,
    "Balanceado + umbral 0.3": res_bal_umbral
}).drop("CM")

print("Comparativa de métricas:\n")
display(df_comp)

# --- Mostrar matrices de confusión ---
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
titulos = ["Original", "Balanceado", "Balanceado + umbral 0.3"]
for ax, res, titulo in zip(axes, [res_original, res_bal, res_bal_umbral], titulos):
    sns.heatmap(res["CM"], annot=True, fmt="d", cmap="Blues", cbar=False, ax=ax)
    ax.set_title(titulo)
    ax.set_xlabel("Predicho")
    ax.set_ylabel("Real")
plt.tight_layout()
plt.show()
```
